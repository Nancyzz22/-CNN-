```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
file_path='F:/math/mathorcup/2023mathorcup/datafeatures1.csv'#这是题目给的图片提取出的特征路径
data = pd.read_csv(file_path,skiprows=0)
features = data.to_numpy()
labels = np.array([1] * 266 + [0] * 35)
print(features)
```

```python
# 将数据集划分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 创建随机森林分类器并设置参数
clf = RandomForestClassifier(n_estimators=100,max_depth=50, random_state=42)
# 使用训练集训练模型
clf.fit(X_train, y_train)
```

```python
# 使用测试集计算正确率
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
with open("F:/math/mathorcup/2023mathorcup/random_forest_model1.pkl", "wb") as f:#保存模型1
    pickle.dump(clf, f)
```
