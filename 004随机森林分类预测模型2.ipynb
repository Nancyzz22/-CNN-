```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
import pickle
file_path='F:/math/mathorcup/2023mathorcup/datafeatures2.csv'#这是数据增强后提取的特征文件datafeatures2.csv路径
data = pd.read_csv(file_path,skiprows=0)
features = data.to_numpy()
labels = np.array([1] * 3990 + [0] * 525)
print(features)
```

```python
# 划分训练集测试集
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.1, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 创建随机森林分类器
clf = RandomForestClassifier(n_estimators=100,max_depth=None, random_state=42)
# 训练模型
clf.fit(X_train, y_train)
importances = clf.feature_importances_
importances_sort = np.argsort(importances)
importances=importances_sort[-200:]
print('前200个重要特征：',importances)
```

```python
# 将特征重要性和重要特征转换为DataFrame
feature_importances = pd.DataFrame({'特征': importances_sort[-200:], '重要性': importances[importances_sort[-200:]]})
print(feature_importances)
```

```python
#重新训练随机森林分类模型：
clf_new = RandomForestClassifier(n_estimators=100,max_depth=None, random_state=40)
clf_new.fit(X_train[:, importances_sort[-200:]], y_train)
```

```python
# 使用测试集评估模型性能
y_pred = clf_new.predict(X_test[:, importances_sort[-200:]])
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#保存模型2
with open("F:/math/mathorcup/2023mathorcup/random_forest_model2.pkl", "wb") as f:
    pickle.dump(clf_new, f)
```
