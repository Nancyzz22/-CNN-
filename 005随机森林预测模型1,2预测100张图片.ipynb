```python
import pandas as pd
import pickle
def compare_arrays(input_array):
    predefined_array = [0]*50 + [1]*50
    n = 0
    for i in range(len(input_array)):
        if input_array[i] == predefined_array[i]:
            n += 1

    return (n+1) / 100
```

```python
with open("F:/math/mathorcup/2023mathorcup/random_forest_model1.pkl", "rb") as f:#这是模型1random_forest_model1.pkl的路径
    loaded_clf = pickle.load(f)
```

```python
file_path='F:/math/mathorcup/2023mathorcup/testfeatures3.csv'#这是100张随机图片特征testfeatures3.csv路径
data = pd.read_csv(file_path,skiprows=0)
features = data.to_numpy()
predictions = loaded_clf.predict(features)
print("Prediction:", predictions)
Accuracy=compare_arrays(predictions)
print('Accuracy:',Accuracy)
```
